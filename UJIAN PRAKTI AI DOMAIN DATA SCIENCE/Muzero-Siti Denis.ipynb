```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

Modelling menggunakan Random Forest

```python
X = data.drop(['IPM'],axis = 1)
y = data['IPM']
```

Menghapus variabel IPM, yang nantinya variabel IPM diinisialisasi sebagai variabel y.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.27, random_state = 48)
X_train.shape, X_test.shape
```

Membagi data training dan testing dengan rasio 27% data digunakan sebagai data testing, dengan pemilihan random state sebesar 48 hingga diperoleh
dimensi dari data trainingnya sebanyak 1603 dan data testingnya yaitu 593

## Modelling

```python
classifier_rf = RandomForestClassifier(n_estimators=15, criterion="entropy")
classifier_rf.fit(X_train, y_train)
```

Modelling menggunakan Random Forest dengan Decision Tree sebanyak 15 dan kriterianya menggunakan entropy untuk perolehan informasi

```python
y_pred = classifier_rf.predict(X_test)
```

Menginisialkan hasil prediksi random forest dengan y_pred dan hasilnya akan digunakan untuk menghitung confussion matriks

## Evaluation

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
print(cm)
```

Menampilkan Confussion matriks 

```python
print('Training-set accuracy score:', classifier_rf.score(X_train, y_train))
```

Menampilkan nilai akurasi untuk prediksi data training yang tinggi yaitu diperoleh sebesar 0.99 atau 99%

```python
print('Testing-set accuracy score:', classifier_rf.score(X_test, y_test))
```

Menampilkan nilai akurasi untuk prediksi data testing yaitu diperoleh sebesar 0.96 atau 96%

# BALANCING DATA

```python
from imblearn.over_sampling import SMOTE
SMOTE = SMOTE()

X_train_SMOTE, y_train_SMOTE = SMOTE.fit_resample(X_train, y_train)
```

Dari data training, dilakukan SMOTE untuk balancing data.

```python
classifier_rf2 = RandomForestClassifier(n_estimators=15, criterion="entropy")
classifier_rf2.fit(X_train_SMOTE, y_train_SMOTE)
```

Menggunakan Random Forest kembali dengan classifier_rf2 karena classifier telah digunakan sebelumnya.
Decision Tree sebanyak 15 dan tetap dengan entropy. Dataset yang digunakan yaitu data yang telah dilakukan SMOTE.

```python
y_pred = classifier_rf2.predict(X_test)
```

Memperoleh nilai y_pred dari variabel X_test menggunakan data yang telah dilakukan SMOTE

```python
cm = confusion_matrix(y_test, y_pred)
```

Menghitung confusion matrix dari y_pred dan y nilai asli

```python
print('Training-set accuracy score:', classifier_rf2.score(X_train_SMOTE, y_train_SMOTE))
print('Testing-set accuracy score:', classifier_rf2.score(X_test, y_test))
```

Menampilkan nilai akurasi untuk prediksi data training yang telah di SMOTE sebesar 99% 
dan prediksi data testing yang telah di SMOTE yaitu 95%
Kedua nilai akurasi tetap tinggi, namun datanya telah balance.

## Model Adaboost

```python
from sklearn.ensemble import AdaBoostClassifier
ABclassifier = AdaBoostClassifier(n_estimators=100)

ABclassifier.fit(X_train_SMOTE, y_train_SMOTE)
y_pred_AB = ABclassifier.predict(X_test)
```

Menambahkan model Adaboost dengan menggunakan data training yang telah di SMOTE.
Diinisialisasikan dengan ABclassifier dan menggunakan 100 decision tree

```python
print('Training-set accuracy score:', ABclassifier.score(X_train_SMOTE, y_train_SMOTE))
print('Training-set accuracy score:', ABclassifier.score(X_test, y_test))
```

Diperoleh nilai akurasi sebesar 38% untuk data training, sedangkan untuk data testing diperoleh akurasi sebesar 15%

## Kesimpulan

Berdasarkan hasil dari perbandingan antara Model Random Forest dan Random Adaboost pada dataset IPM ini,
Model Random Forest ini menunjukkan hasil akurasi yang lebih baik dari pada Model Adaboost, yaitu nilai akurasi sebesar 99%.
Ini membuktikan bahwa model yang sesuai dengan dataset IPM ialah Model Random Forest dengan menggunakan parameter n_estimators=15 dan criterion='entropy'.
